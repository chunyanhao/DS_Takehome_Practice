###  2. Summerize each company's daily employees count

```python
# 1. create data frame for each date and company
date_df = pd.DataFrame(pd.date_range(start='2011-01-24', end='2015-12-13'), columns = ['date'])
company_df = pd.DataFrame(data['company_id'].unique(), columns = ['company_id']).sort_values(by = ['company_id'])
part1_data = pd.merge(left = date_df, right = company_df, how = 'cross')
part1_data.head()
```

```python
# 2. calculate how many employees join and quit in each company everyday
daily_join = data[['employee_id', 'company_id', 'join_date']].groupby(['join_date', 'company_id']).count().reset_index()
daily_quit = data[['employee_id', 'company_id', 'quit_date']].groupby(['quit_date', 'company_id']).count().reset_index()
```

```python
# 3. merge dataframe with join and quit data
daily_report = pd.merge(left = part1_data, 
                      right = daily_join, 
                      how = 'left', 
                      left_on = ['date', 'company_id'],
                      right_on = ['join_date', 'company_id'])
daily_report = pd.merge(left = daily_report, 
                      right = daily_quit, 
                      how = 'left', 
                      left_on = ['date', 'company_id'],
                      right_on = ['quit_date', 'company_id'])
```

```python
# revise the table format and calculate the daily headcount.
daily_report.columns = ['date', 'company_id', 'join_date', 'join_employee_count',
                      'quit_date', 'quit_employee_count']
daily_report = daily_report[['date', 'company_id', 'join_employee_count','quit_employee_count' ]]

daily_report['quit_employee_count'] = daily_report['quit_employee_count'].replace(np.nan, 0)
daily_report['join_employee_count'] = daily_report['join_employee_count'].replace(np.nan, 0)

daily_report['net_add'] = daily_report['join_employee_count'] - daily_report['quit_employee_count']
daily_report['employee_headcount'] = daily_report.groupby('company_id')['net_add'].transform(pd.Series.cumsum)
```

### 3. Explore the main factors that drive employee churn

### 1) quit vs employee days

```python
data['quit_year'] = (data['quit_date'] - data['join_date']).dt.days/365
```

```python
plt.figure(figsize = (12, 4))
sns.histplot(data = data, x = 'quit_year')
plt.show()
```

The histogram above shows that after people worked in the company for 1 year, they are more likely to resign.

Besides, the more time they stay in the company, the less likely they tend to leave.

### 2) quit vs quit_date(month and week)

```python
data['quit_month'] = data['quit_date'].dt.month
# plt.figure(figsize = (20, 4))
sns.catplot(data = data, x = 'quit_month',kind = 'count', height=2, aspect=4)
plt.show()
```

```python
data['quit_week'] = data['quit_date'].dt.week

sns.catplot(data = data, x = 'quit_week',kind = 'count', height=4, aspect=8)
plt.show()
```

The histograms by week and month shows that people have a high preference to resign in October, November and May.

### 3) quit vs seniority

```python
data['quit_seniority']  = round(data['quit_year'],0) + data['seniority']
```

```python
plt.figure(figsize = (12, 4))
sns.histplot(data = data, x = 'quit_seniority')
plt.show()
```

The histograms by seniority before people quit shows that:

    1) working experiences less than 18 years are more likely to quit;
    2) After accumulating 1 year working experience, people will decide to leave immediately, or will stay at least 2 years.
    3) People who have more than 20 years working experience are less likely to resign.

### 4) quit vs salary

```python
data['quit'] = data['quit_date'].isna()
plt.figure(figsize = (12, 4))

sns.histplot(data = data, x = 'salary', bins = 40, hue = 'quit')
plt.show()
```

we explore the relationship of salary and quit, and we find that:

    1) The salary distribution among employee who leave and stay are similar, mainly concentrated in a range of [60,000, 140,000] The number of employee who quit the job are less than those stay.
    2) Employees with a salary range of [60,000, 120,000] are more likely to quit. 

## 4. Verify the explore conclusion with a decision tree model

```python
data.columns
```

```python
data['quit_date'] = data['quit_date'].replace(np.nan, '2015-12-13')
data['quit_year'] = (data['quit_date'] - data['join_date']).dt.days/365
data['quit_seniority']  = round(data['quit_year'],0) + data['seniority']

reg_data = data[['company_id', 'dept', 'seniority','quit_seniority', 'salary','quit_year', 
                 'quit_month', 'quit_week', 'quit']]
reg_data = pd.get_dummies(reg_data, 
                          columns = ['company_id', 'dept', 'quit_month', 
                                     'quit_week','quit'])
```

```python
X_train = reg_data.iloc[:,:-2]
y_train = reg_data['quit_True']

fullClassTree = DecisionTreeClassifier(random_state=1, max_depth = 5)
fullClassTree.fit(X_train, y_train)
```

```python
from dmba import plotDecisionTree

plotDecisionTree(fullClassTree, feature_names = X_train.columns)
```

```python
# for importance, name in sorted(zip(fullClassTree.feature_importances_, X_train.columns),reverse=True):
#     print (name, importance)
```

The Tree plotted above indicates that the most important features affecting the employee quit or not are:
    1) how many year have they stayed in this company
    2) Which month will they decide to leave.
except for the leaving timeline, we can also do some further exploratory analysis work on other features: 

```python
reg2_data = data[['company_id', 'dept', 'seniority', 'quit_seniority', 'salary', 'quit']]
reg2_data = pd.get_dummies(reg2_data, columns = ['company_id', 'dept','quit'])

X_train = reg2_data.iloc[:,:-2]
y_train = reg2_data['quit_True']

ClassTree_2 = DecisionTreeClassifier(random_state=1, max_depth = 5)
ClassTree_2.fit(X_train, y_train)

plotDecisionTree(ClassTree_2, feature_names = X_train.columns)
```

```python
for importance, name in sorted(zip(ClassTree_2.feature_importances_, X_train.columns),reverse=True):
    print (name, importance)
```

The Feature importance displayed above indicated that except for the leaving timeline, year of working experience and salary are also very important for employees to decide leaves the company or not.

## 3. Question 2

Build a model to optimize in future email campaigns to maximize the probability of users clicking on the link inside the email

### 3.1 Model fitting

```python
y = data['clicked']
x = pd.get_dummies(data.iloc[:, 1:7], 
                   columns = ['email_text','email_version', 'weekday', 
                               'user_country'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from dmba import plotDecisionTree

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix, recall_score, precision_score, accuracy_score
from sklearn.metrics import roc_curve, roc_auc_score, auc
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
ClassTree = DecisionTreeClassifier(random_state=1, max_depth = 5)
ClassTree.fit(x_train, y_train)
```

```python
plotDecisionTree(ClassTree, feature_names = x_train.columns)
```

```python
rf_reg_importances = pd.DataFrame(ClassTree.feature_importances_, index = x_test.columns)

rf_reg_importances.columns = ['importance']
rf_reg_importances.sort_values(by = 'importance', ascending = False)[:8]
```

The Regression tree displayed about indicates that the number of items users bought before, version of email and features like country, hour and weekday can affect weather the user click the linked in the email or not.

```python
rf_reg = RandomForestClassifier(max_features = 4)
rf_reg.fit(x_train, y_train)
```

```python
model = rf_reg
train_con_matrix = confusion_matrix(y_train, model.predict(x_train))
print('train matrix: \n', train_con_matrix)
print('train accuracy:', round(accuracy_score(y_train, model.predict(x_train)),4))

test_con_matrix = confusion_matrix(y_test, model.predict(x_test))
print('\n test matrix: \n', test_con_matrix)
print('test accuracy:', round(accuracy_score(y_test, model.predict(x_test)), 4))
```

```python
rf_reg_importances = pd.DataFrame(rf_reg.feature_importances_, index = x_test.columns)

rf_reg_importances.columns = ['importance']
rf_reg_importances.sort_values(by = 'importance', ascending = False)[:8]
```

The result of random forest also emphasize on features like hour and he number of items users bought before.

It can also be helpful by generating some plots to specify how dose these features affect the users click the link or not.

### 3.2 Visualization

```python
fig, ax = plt.subplots(3, 1, figsize = (12,12))
sns.barplot(data = data, x = 'hour', y = 'clicked', estimator = np.mean, ax = ax[0])
order = ['Monday','Tuesday', 'Wednesday', 'Thursday','Friday','Saturday', 'Sunday']
sns.barplot(data = data, x = 'weekday', y = 'clicked', estimator = np.mean, ax = ax[1], order = order)
sns.barplot(data = data, x = 'user_past_purchases', y = 'clicked', estimator = np.mean, ax = ax[2])

plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15,4))
sns.barplot(data = data, x = 'email_text', y = 'clicked', estimator = np.mean, ax = ax[0])
sns.barplot(data = data, x = 'email_version', y = 'clicked', estimator = np.mean, ax = ax[1])
sns.barplot(data = data, x = 'user_country', y = 'clicked', estimator = np.mean, ax = ax[2])
plt.show()
```

Combine with the data visualization, we can conclude that:

    1) Users with a high level of item purchase history are more likely to click the email link. This is reasonable that those loyal customers are more interested in the promotion emails.
    
    2) Sending email with a personalized formate can result in a higher level of click rate, because it makes users feel more comfortable and valuable.
    
    3) The website has more effective users in US and UK compared with FR and ES.
    
    4) The best time to send out the email is before the mid-night.
    
    5) People are more likely to click the link during weekdays instead of weekends. The reason may be that during weekends leisure time, people do not want to check the mail box at all.
